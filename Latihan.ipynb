# Latihan Grafik

## Pertumbuhan Kucing
Kamu diminta membuat grafik pertumbuhan seekor anak kucing dalam 6 minggu. Berikut adalah datanya:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,6])
y = np.array([0.5,0.7,1.0,1.3,1.5,1.8])

plt.plot(x,y, 'ro-')
plt.xticks(np.arange(1, 11, 1))
plt.yticks(np.arange(0.0, 2.1, 0.2))

plt.title('Pertumbuhan berat anak kucing', fontdict={'fontname': 'Comic Sans MS'})
plt.xlabel('Minggu', fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('Berat (kg)', fontdict={'fontname': 'Comic Sans MS'})

plt.show()
```

## Penjualan Toko Roti

```python
hari = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']
terjual = np.array([12,18,25,22,30,40,35])

plt.plot(hari, terjual, color='green', marker='o')
plt.yticks(np.arange(0, 51, 5))
plt.grid(axis='x', linestyle=':')
plt.grid(axis='y', linestyle=':')

plt.title('Penjualan Roti')
plt.xlabel('Hari')
plt.ylabel('Terjual')

for i in range(len(hari)):
    plt.text(hari[i], terjual[i] + 2, str(terjual[i]), ha='center', fontsize=10)

plt.show()
```
